```python
import cv2
#imports cv library
```

```python
import os
```

```python
img = cv2.imread('happy (3).jpg')
#image path
```

```python
from deepface import DeepFace
#imports deep face library
```

```python
import matplotlib.pyplot as plt
#imports matplotlib which is used to display the images on a plane
```

```python
plt.imshow(img)
#displays the image
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#changes the photo to its default state from RGB
```

```python
predictions = DeepFace.analyze(img)
#used deep face library to analyze and make predictions form the image
```

```python
predictions
#displays all the predictions 
```

```python
type(predictions)
```

```python
first_prediction = predictions[0]
```

```python
dominant_emotion = first_prediction['dominant_emotion']
```

```python
print("Dorminant Emotion:", dominant_emotion)
```

#Rectangle around the face or bounds

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#haarcascade file used draw a square around the face 
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray, 1.1, 4)
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    #dimensions for the square
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#display the square arounf the image
```

```python
first_prediction = predictions[0]
```

```python
dominant_emotion = first_prediction['dominant_emotion']
```

```python
dominant_emotion_str = str(dominant_emotion)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, dominant_emotion_str, (0, 50), font, 1, (0, 0, 255), 2, cv2.LINE_4)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
